```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
d = 200 # grid points

t_left = 0
t_right = 1
t = np.linspace(t_left,t_right,d)
```

$$ w^\ast (t) = \sum_{j=1}^{M+1}  \frac{\sin(2 \pi j t )}{j^\beta} $$

```python
def sin_polin(t,alpha,M):
    sum = 0
    for j in range(1,M+2):
        sum += np.sin(2*j*np.pi*t) / (j**alpha)
    return sum
```

```python
beta = .4
M = 10

true_w = sin_polin(t,beta,M)                        


plt.plot(t,true_w)
```
